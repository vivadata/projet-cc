```python
import pandas as pd
import csv
```

## Fonction lecture, extraction et enregistrement csv

```python


def extraire_csv(file_in, file_out,
                 ligne_colonnes=60,
                 ligne_donnees=62):
    lignes = []

    # Lecture de toutes les lignes
    with open(file_in, "r", encoding="utf-8") as f:
        lignes = f.readlines()

    # Ligne 60 → index 59
    ligne_col = lignes[ligne_colonnes - 1].strip()
    if ligne_col.startswith("# "):
        ligne_col = ligne_col[2:]

    # Split des colonnes
    colonnes = ligne_col.rstrip(";").split(";")

    # --- Extraction des données ---
    data = []
    for ligne in lignes[ligne_donnees - 1:]:
        ligne = ligne.strip()
        if ligne == "":
            continue
        valeurs = ligne.rstrip(";").split(";")
        data.append(valeurs)

    # --- Écriture du CSV ---
    with open(file_out, "w", newline="", encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile, delimiter=";")
        writer.writerow(colonnes)
        writer.writerows(data)

    print(f"Extraction terminée. CSV écrit dans : {file_out}")
```

## Extraction et test RWL29

```python
filepath = '/content/drive/MyDrive/Projet CC/Data manipulations/Simu data/simu_RWL29.txt'
with open(filepath, "r") as file:
  lignes = []
  for index, ligne in enumerate(file, start=1):
        lignes.append(ligne.rstrip("\n"))

print(lignes[59])
print(lignes[61])
```

```python
file_in = '/content/drive/MyDrive/Projet CC/Data manipulations/Simu data/simu_RWL29.txt'
file_out = '/content/drive/MyDrive/Projet CC/Data manipulations/Simu data/simu_RWL29.csv'
extraire_csv(
    file_in = file_in,
    file_out = file_out
)
```

```python
df = pd.read_csv(file_out, delimiter = ';')
df.head()
```

## Extraction des infos de la grille BRIO et enregistrement dans un fichier points.csv
