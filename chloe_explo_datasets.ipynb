## Extraction des infos contenues dans les fichiers 'Données climatologiques'

```python
#imports
import os
import pandas as pd
from datetime import datetime
```

### Données climatologiques de base - 6 minutes

```python
### Données climatologiques de base - data urls
urls_quot_rr_t_vent = ['https://www.data.gouv.fr/api/1/datasets/r/ca71c620-8310-4957-86c4-7a2cb827edf9',
                       'https://www.data.gouv.fr/api/1/datasets/r/fe8b9ae4-885c-4a12-98b4-af3ec330103f',
                       'https://www.data.gouv.fr/api/1/datasets/r/565ce1ce-7cd5-4a9b-a208-4c3bd52b927e']


urls_quot_autres = ['https://www.data.gouv.fr/api/1/datasets/r/a22d8046-8eaf-4a1d-9996-b9fe0630b555',
                    'https://www.data.gouv.fr/api/1/datasets/r/1c4c8fb0-1e28-4349-bd7e-6d47939da21f']

urls_decad_agro = ['https://www.data.gouv.fr/api/1/datasets/r/3d7c7887-2278-42fa-a64b-48d5eab7ff21',
                   'https://www.data.gouv.fr/api/1/datasets/r/feb0d199-45f2-4ca7-bea6-45cd5e85bc01',
                   'https://www.data.gouv.fr/api/1/datasets/r/3da7e0b0-4e90-4272-a909-2fe0b8727309']
```

```python
df = pd.DataFrame()
for url in urls_quot_rr_t_vent:
  df_new = pd.read_csv(url, compression='gzip', delimiter=';')
  df = pd.concat([df,df_new])
  print(url)
```

```python
#Quality flag columns
flag_col = [col for col in df.columns if col.startswith('Q')]

#Removing NAs and duplicates
df_cleaned = df.dropna(axis= 'columns', how='all')
df_cleaned = df_cleaned.drop_duplicates()

#removing rows where Q == 2
#df_cleaned = df_cleaned.drop(df_cleaned[df_cleaned[flag_col] == 2].index)
```

```python
df.shape, df_cleaned.shape
```

```python
#save to csv
filepath = os.path.join('data','RAW_QUOT_rr-T-vent_1900-2025.csv')
df_cleaned.to_csv(filepath, index=False)
```
